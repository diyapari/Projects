```python
# input variable
X = BC_df.drop(['target'], axis = 1)
X.head()
```

```python
# output variable
y = BC_df['target']
y.head()
```

```python
# split dataset into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state= 5)
X_train
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
# Support vector classifier
#Testing accuracy
from sklearn import svm
svc = svm.SVC()
svc.fit(X_train, y_train)
y_pred_sv = svc.predict(X_test)
print(accuracy_score(y_test,y_pred_sv))

# Train with Standard scaled Data
svc2= svm.SVC()
svc2.fit(X_train_sc, y_train)
y_pred_svc_sc = svc2.predict(X_test_sc)
accuracy_score(y_test, y_pred_svc_sc)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
lrc = LogisticRegression(random_state = 51)
lrc.fit(X_train, y_train)
y_pred_lr = lrc.predict(X_test)
print(accuracy_score(y_test, y_pred_lr))

# Train with Standard scaled Data
lr2 = LogisticRegression(random_state = 51)
lr2.fit(X_train_sc, y_train)
y_pred_lr_sc = lr2.predict(X_test_sc)
accuracy_score(y_test, y_pred_lr_sc)
```

```python
# K – Nearest Neighbor Classifier
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
print(accuracy_score(y_test, y_pred_knn))

# Train with Standard scaled Data
knn2 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
knn2.fit(X_train_sc, y_train)
y_pred_knn_sc = knn2.predict(X_test_sc)
accuracy_score(y_test, y_pred_knn_sc)
```

```python
# Naive Bayes Classifier
from sklearn.naive_bayes import GaussianNB
nbc = GaussianNB()
nbc.fit(X_train, y_train)
y_pred_nb = nbc.predict(X_test)
print(accuracy_score(y_test, y_pred_nb))

# Train with Standard scaled Data
nb2 = GaussianNB()
nb2.fit(X_train_sc, y_train)
y_pred_nb_sc = nb2.predict(X_test_sc)
accuracy_score(y_test, y_pred_nb_sc)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dtc.fit(X_train, y_train)
y_pred_dt = dtc.predict(X_test)
print(accuracy_score(y_test, y_pred_dt))

# Train with Standard scaled Data
dt2 = DecisionTreeClassifier(criterion = 'entropy', random_state = 51)
dt2.fit(X_train_sc, y_train)
y_pred_dt_sc = dt2.predict(X_test_sc)
accuracy_score(y_test, y_pred_dt_sc)
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
print(accuracy_score(y_test, y_pred_rf))

# Train with Standard scaled Data
rf2 = RandomForestClassifier(n_estimators = 20, criterion = 'entropy', random_state = 51)
rf2.fit(X_train_sc, y_train)
y_pred_rf_sc = rf2.predict(X_test_sc)
accuracy_score(y_test, y_pred_rf_sc)
```

```python
# Adaboost Classifier
from sklearn.ensemble import AdaBoostClassifier
adb_classifier = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'entropy', random_state = 200),
                                    n_estimators=2000,
                                    learning_rate=0.1,
                                    algorithm='SAMME.R',
                                    random_state=1,)
adb_classifier.fit(X_train, y_train)
y_pred_adb = adb_classifier.predict(X_test)
print(accuracy_score(y_test, y_pred_adb))

# Train with Standard scaled Data
adb2 = AdaBoostClassifier(DecisionTreeClassifier(criterion = 'entropy', random_state = 200),
                                    n_estimators=2000,
                                    learning_rate=0.1,
                                    algorithm='SAMME.R',
                                    random_state=1,)
adb2.fit(X_train_sc, y_train)
y_pred_adb_sc = adb2.predict(X_test_sc)
accuracy_score(y_test, y_pred_adb_sc)
```

```python
# XGBoost Classifier
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train, y_train)
y_pred_xgb = xgb.predict(X_test)
print(accuracy_score(y_test, y_pred_xgb))

# Train with Standard scaled Data
xgb2 = XGBClassifier()
xgb2.fit(X_train_sc, y_train)
y_pred_xgb_sc = xgb2.predict(X_test_sc)
accuracy_score(y_test, y_pred_xgb_sc)
```

```python
print("Accuracy of Spport vector machine classifier:",accuracy_score(y_test, y_pred_sv))
print("Accuracy of Logistic regression classifier :",accuracy_score(y_test, y_pred_lr))
print("Accuracy of K- Nearest Neighbour classifier:",accuracy_score(y_test, y_pred_knn))
print("Accuracy of Naive Bayes classifier:",accuracy_score(y_test, y_pred_nb))
print("Accuracy of Decision Tree classifier:",accuracy_score(y_test, y_pred_dt))
print("Accuracy of Random Forest classifier:",accuracy_score(y_test, y_pred_rf))
print("Accuracy of Adaboost classifier:",accuracy_score(y_test, y_pred_adb))
print("Accuracy of Xgboost classifier:",accuracy_score(y_test, y_pred_xgb))
```

```python
# Cross validation
from sklearn.model_selection import cross_val_score
cross_validation = cross_val_score(estimator = xgb, X = X_train,y = y_train, cv = 10)
print("Cross validation accuracy of XGBoost model = ", cross_validation)
print("\nCross validation mean accuracy of XGBoost model = ", cross_validation.mean())
```

```python
cm = confusion_matrix(y_test, y_pred_xgb)
plt.title('Heatmap of Confusion Matrix', fontsize = 10)
sns.heatmap(cm, annot = True)
plt.show()
```
